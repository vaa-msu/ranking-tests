## Двоенко, Пшеничный (2015)

Двоенко, С.Д., Пшеничный, Д.О. **О метрических свойствах медианы Кемени.** *Машинное обучение и анализ данных,* т.1, №11, 1619--1631, 2015. ([PDF](http://jmlda.org/papers/doc/2015/no11/Dvoenko2015KemenyMedian.pdf))

Из-за ограничений в анализаторе ранжирований пакета `Rankings.jl` альтернативы должны следовать в алфавитном порядке, поэтому их условные имена были образованы из кириллической буквы `П` (проект) и двух нижних индексов: `П₀₁`,...,`П₁₄`, а ранжирования записаны в соответствии с (немного исправленной -- см. ниже) таблицей из статьи.

```julia
Dvoenko2015 = ranking"П₀₂,П₁₁,П₀₃,П₁₂,П₀₁,П₀₆,П₁₃,П₀₅,П₀₄,П₀₇,П₀₈,П₁₀,П₀₉,П₁₄"⦸
              ranking"П₀₂,П₁₁,П₀₃,П₀₁,П₀₇,П₀₆,П₀₄,П₁₂,(П₀₅,П₁₃),(П₀₈,П₀₉,П₁₀,П₁₄)"⦸
              ranking"П₁₀,П₁₁,(П₀₂,П₁₄),(П₀₅,П₀₉),П₁₃,(П₀₃,П₀₆),П₀₄,П₀₈,(П₀₇,П₁₂),П₀₁"⦸
              ranking"П₀₈,П₁₁,П₀₃,П₀₂,(П₀₁,П₀₄,П₀₅,П₀₆,П₁₀,П₁₂,П₁₃),П₀₉,(П₀₇,П₁₄)"⦸
              ranking"(П₀₅,П₀₈,П₀₉,П₁₄),(П₀₇,П₁₁,П₁₃),(П₀₄,П₀₆,П₁₂),П₀₁,(П₀₂,П₁₀),П₀₃"⦸
              ranking"(П₀₈,П₀₉,П₁₄),(П₀₄,П₀₇),П₁₀,(П₀₅,П₁₁),П₀₆,(П₁₂,П₁₃),П₀₁,П₀₃,П₀₂"⦸
              ranking"(П₀₁,П₀₄,П₀₅,П₀₇,П₀₈,П₀₉,П₁₂,П₁₃),(П₀₂,П₁₄),П₀₃,(П₀₆,П₁₁),П₁₀"⦸
              ranking"П₁₀,П₀₄,П₀₉,П₀₇,(П₀₁,П₁₂),П₀₃,(П₀₈,П₁₄),П₀₂,П₀₅,П₁₃,П₀₆,П₁₁"
```

В статье (стр.1626) приведена таблица, где рассматриваемым альтернативам по результатам
экспертизы присвоены значения рангов, показывающие предпочтительность альтернатив
(наиболее предпочтительная альтернатива имеет ранг 1, следующая за ней -- 2 и т.д.).
Если несколько альтернатив имеют одинаковые ранги, то в таблице указывается некоторый
стандартизированный ранг, равный (как можно понять из приводимых в ней величин)
среднему значению из соседних (различных) значений рангов.

В этой таблице рангов имеется опечатка, поскольку шестое экспертное ранжирование
содержит для альтернативы с номером 3 неправильное значение ранга (3): этот ранг 3
уже вошёл в стандартизированное значение ранга 2 (`(1+2+3)/3 -> 2`); поэтому вместо
этого ранга было использовано «выпавшее» из столбца значение ранга 13.

Тем не менее матрица штрафов, присутствующая на той же странице 1626, ошибок не содержит
и соответствует используемой здесь «облегчённой» матрице потерь для приводимого там
набора ранжирований. В этом легко убедиться, перейдя от «облегчённой» матрицы потерь к традиционной при помощи матрицы `J(N) = [0+(i!=j) for i=1:N, j=1:N]`, добавляемой к «облегчённой» с коэффициентом 8:

```julia
J(N) = [0+(i!=j) for i=1:N, j=1:N];
Dvoenko2015 + 8J(14)
```

Однако указаное там для медианы Кемени итоговое ранжирование альтернатив (по их номерам):
`[11,4,8,2,5,9,7,12,1,3,13,14,6,10]` вовсе не является оптимальным (сумма наддиагональных
элементов «облегчённой» матрицы для него равна -110), ибо даже циклический сдвиг первых
трёх номеров вправо даёт ранжирование `[8,11,4,2,5,9,7,12,1,3,13,14,6,10]` с суммой
наддиагональных элементов -112. И вообще, как оказывается, всего ранжирований с такой
«улучшенной» наддиагональной суммой имеется целых 160.

```julia
Rankings.upper_sum(Dvoenko2015[[11,4,8,2,5,9,7,12,1,3,13,14,6,10],[11,4,8,2,5,9,7,12,1,3,13,14,6,10]])
```
